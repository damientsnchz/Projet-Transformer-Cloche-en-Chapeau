```python
def generate_spade_points(N):
    # Paramètres du pique
    length = 2.5  # Longueur de la pointe
    width = 3.0   # Largeur de la pointe
    curve_height = 1.5  # Hauteur de la courbe
    curve_width = 0.8   # Largeur de la courbe

    # Générer les points pour la pointe
    x_tip = np.linspace(-width/2, width/2, N)
    y_tip = -np.sqrt((length**2)*(1 - (x_tip**2)/(width**2)))

    # Générer les points pour la courbe
    x_curve = curve_width * np.sin(np.linspace(-np.pi/2, np.pi/2, N))
    y_curve = curve_height * np.cos(np.linspace(-np.pi/2, np.pi/2, N))

    # Combinez les points de la pointe et de la courbe
    x_spade = np.concatenate([x_tip, x_curve])
    y_spade = np.concatenate([y_tip, y_curve])

    return x_spade, y_spade

# Générer les coordonnées des points pour le pique
N = 100
x_spade, y_spade = generate_spade_points(N)

# Tracer le pique
plt.figure(figsize=(8, 6))
plt.plot(x_spade, y_spade, 'k-')
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Pique')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

```
